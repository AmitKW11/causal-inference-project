```python
#reminder for self:
#propensity scores: The probability of receiving treatment given the observed confounders
#P(T=1 | X) = probability someone gets treated based on their X values
```

```python
from sklearn.metrics import brier_score_loss
from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt


def train_propensity_model(model, X_train, T_train):
    """
    Return a fitted propensity model on your training data
    :param model: 
    :param X_train: 
    :param T_train: 
    :return: 
    """
    return model.fit(X_train, T_train)

def eval_propensity_model(fitted_model, X_val, T_val):
    """
    Check your fitted propensity model using the brier score and calibration curve.
    
    You must print or return (and then print for us to see) the brier score. 
    You must show the calibration curve.
    :param fitted_model: 
    :param X_val: 
    :param T_val: 
    :return: 
    """
    propensity_scores = fitted_model.predict_proba(X_val)[:, 1]
    
    brier = brier_score_loss(T_val, propensity_scores)
    print(f"Brier Score: {brier:.4f}")
    
    fraction_of_positives, mean_predicted_value = calibration_curve(T_val, propensity_scores, n_bins=10)
    
    plt.figure(figsize=(8, 6))
    plt.plot(mean_predicted_value, fraction_of_positives, "s-", label="Model")
    plt.plot([0, 1], [0, 1], "k--", label="Perfect calibration")
    plt.xlabel("Mean Predicted Probability")
    plt.ylabel("Fraction of Positives")
    plt.title("Calibration Curve")
    plt.legend()
    plt.grid(True)
    plt.show()
    return brier
```

### Q1. 
Explain what the Brier score represents? What is the best possible Brier score? What is the Brier score if the model always guesses 0.5?\
### A1:
The Brier score measures how well a model's probability predictions align with actual binary outcomes. It calculates the average squared difference between predicted probabilities and true results. Each predicted probability $p_i$ and the actual outcome $o_i \in \{0,1\}$. Because it is effectively an MSE on probabilities, **lower values indicate better calibration and discrimination**. The *best possible* score is **0**, achieved only by a model that assigns $p_i = 1$ whenever the event occurs and $p_i = 0$ whenever it does not. By contrast, a coin-flip model that always predicts $p_i = 0.5$ yields a constant squared error of $(0.5 - 1)^2 = (0.5 - 0)^2 = 0.25$ for every observation, so its overall Brier score is **0.25**, which is bascially a random model


### Q2.
Explain what the calibration plot shows? What is the best possible calibration plot?  
### A2: 
The calibration plot displays the relationship between predicted probabilities (x-axis) and actual treatment rates (y-axis)
by grouping observations into bins based on their predicted probabilities.
For each bin, it shows whether the model's predicted probabilities match the observed frequencies of treatment assignment.
The best possible calibration plot would be a perfect diagonal line from (0,0) to (1,1), where every predicted probability exactly matches the observed treatment frequency. This would mean that when the model predicts 30% treatment probability, exactly 30% of individuals in that group actually receive treatment, and so forth across all probability levels. The deviations from this diagonal in the current plot indicate calibration errors that could impact the reliability of propensity scores for causal inference.

### Q3.
The sklearn implementation of calibration_plot (see Tutorial) has a parameter "n_bins". Explain this parameter. What happens if you use too many bins (for example 100 on this dataset) or too little bins (for example 1)?
### A3:  
The n_bins parameter divides predicted probabilities into groups for calibration analysis. Each bin calculates the mean predicted probability versus actual treatment rate to create the calibration curve.

Too many bins (100) creates statistical unreliability because most bins contain too few observations, producing noiic calibration plots where randobehaviourns mask true model performance. A bin with only 1-2 people might show 100% treatment rate pureby chance.

Too few bins (1) eliminavariationrof the datament. With one bin, all data collapses into a single point showing overall treatment rate versus mean prediction, providing no information about calibration patterns across different probability ra - which is what we wanted to get from the plot in the first place size.

```python
import matplotlib.pyplot as plt

def check_overlap(propensity_model, X_train, T_train):
    """
    Show histograms of the propensity on the T=0 and the T=1 group like in the tutorial
    Use this to ascertain if there is a violation of overlap.
    
    You must show the graph.
    :param propensity_model: 
    :param X_train: 
    :param T_train: 
    :return: 
    """
    # Step 1: Get propensity scores (probabilities of treatment)
    T_prob_train = propensity_model.predict_proba(X_train)[:, 1]
    
    # Step 2: Separate probabilities by treatment group
    prob_treatment_0 = T_prob_train[T_train == 0]
    prob_treatment_1 = T_prob_train[T_train == 1]
    
    # Step 3: Create histograms
    plt.figure(figsize=(10, 6))
    plt.hist(prob_treatment_0, bins=20, alpha=0.6, label='Treatment = 0', color='blue')
    plt.hist(prob_treatment_1, bins=20, alpha=0.6, label='Treatment = 1', color='red')
    
    # Step 4: Add labels and formatting
    plt.xlabel('Predicted Probability')
    plt.ylabel('Frequency')
    plt.title('Histograms of Predicted Probabilities for Treatment = 0 and Treatment = 1')
    plt.legend(loc='upper right')
    plt.grid(True, alpha=0.3)
    
    # Show the plot
    plt.show()
```

## S-Learners

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np

def train_s_learner(model, X_train, y_train):
    """
    Return a fitted s-learner model on your training data
    :param model: 
    :param X_train: 
    :param y_train: 
    :return: 
    """
    return model.fit(X_train, y_train)

def s_learner_model_selection(models, metric, X_train, y_train, X_val, y_val):
    """
    Use the validation set to select the model with the best performance over the provided metric (e.g. rmse)
    Note that X_train and X_val must include the treatment column for the s-learner.
    
    You may select models in a more complicated way (like you would in real life). But here you can simplify.
    You must either return the fitted model or show the scores that lead to your decision.
    :param model: 
    :param metric:
    :param X_train: 
    :param y_train: 
    :param X_val: 
    :param y_val: 
    :return: 
    """
    
    best_score = float('inf')   
    best_model = None
    scores = {}
    
    for name, model in models.items():
        
        fitted_model = train_s_learner(model, X_train, y_train)
        y_pred = fitted_model.predict(X_val)
        
        # Calculate the specified metric
        if metric == 'rmse':
            score = np.sqrt(mean_squared_error(y_val, y_pred)) 
       
        elif metric == 'mae':
            score = mean_absolute_error(y_val, y_pred)
        
        scores[name] = score
        
        if score < best_score:
            best_score = score
            best_model = fitted_model
    
    print("Model Performance:")
    for name, score in scores.items():
        print(f"{name}: {metric.upper()} = {score:.4f}")
    
    return best_model, scores
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor


X_train_with_T = pd.concat([X_train, T_train], axis=1)
X_val_with_T = pd.concat([X_val, T_val], axis=1)


models = {
    'Linear': LinearRegression(),
    'RandomForest': RandomForestRegressor(n_estimators=100, random_state=random_seed),
    'GradientBoosting': GradientBoostingRegressor(n_estimators=100, random_state=random_seed)
}

fitted_s_learner, scores = s_learner_model_selection(
    models=models,
    metric='rmse',
    X_train=X_train_with_T,
    y_train=y_train,
    X_val=X_val_with_T,
    y_val=y_val
)
```

```python
import pandas as pd
import numpy as np

def prepare_data_for_s_learner_inference(X):
    """
    Receives X, data, and returns two sets of data. One with T=1 and the other with T=0.
    For example, if the X = [x1, x_2] then this function returns X1 = [x1, x2, 1] and X0 = [x1, x2, 0]
    
    You must return two datasets, one for T=1 and one for T=0.
    :param X: 
    :return: 
    """
    if isinstance(X, pd.DataFrame):
        # Insert treatment column for DataFrames
        treatment_col = 'T'
        X1 = X.copy()
        X1[treatment_col] = 1
        X0 = X.copy()
        X0[treatment_col] = 0
    elif isinstance(X, np.ndarray):
        X1 = np.concatenate([X, np.ones((X.shape[0], 1))], axis=1)
        X0 = np.concatenate([X, np.zeros((X.shape[0], 1))], axis=1)
    else:
        raise TypeError("Input X must be a pandas DataFrame or numpy array.")
    
    return X1, X0

def calculate_s_learner_ate(s_learner_model, X):
    """
    Receives an s learner model and data
    
    You must return or print the ate estimation on X based on the S-Learner.
    :param s_learner: 
    :param X: 
    :return: 
    """
    X1, X0 = prepare_data_for_s_learner_inference(X)
    
    y1_pred = s_learner_model.predict(X1)  # Predicted outcomes if everyone is treated
    y0_pred = s_learner_model.predict(X0)  # Predicted outcomes if nobody is treated
    
    individual_treatment_effects = y1_pred - y0_pred
    
    ate = individual_treatment_effects.mean()
    
    # Print the results
    print(f"S-Learner Average Treatment Effect (ATE): {ate:.4f}")
    
    return ate
```

### Q6. 
Using a Lasso model, check the ATE over different regularization parameters (see below) \
What happens to the ATE with the S-Learner? Why?

### A6:
As the Lasso regularization parameter increases, the estimated ATE from the S-learner decreases and eventually approaches zero. This happens because Lasso shrinks all coefficients toward zero, including the treatment coefficient. With higher regularization, the model basically suppresses the treatment effect along with everything else.
When the regularization gets too strong, the model can't tell the difference between the treatment effect and random noise, so it treats the treatment variable like any other feature and shrinks its coefficient down. Eventually, the model becomes so conservative that it estimates no treatment effect at all, even if there really is one in the data.
This shows a key problem with using Lasso for causal inference - while regularization helps prevent overfitting, it can also accidentally eliminate the causal effect we're trying to measure. The model ends up being too cautious and underestimates or completely misses real treatment effects.

this was a really interesting concept to learn about.

Add 95% confidence intervals around the point estimation. You have freedom to choose how to do this. For example, bootstraps, parametric methods (e.g. linear regression coefficient), etc.

```python
from sklearn.utils import resample
from sklearn.base import clone

def s_learner_confidence_intervals(s_learner_model, X_train, y_train, X_val, y_val):
    """
    Receives an s-learner model and returns 95% confidence intervals.
    You do not need to use every parameter passed to this method and you can add additional parameters.
    
    You must either print, visualize the confidence intervals or return them as a list.
    :param s_learner_model: 
    :param X_train: 
    :param y_train: 
    :param X_val: 
    :param y_val: 
    :return: 
    """
    bootstrap_ates = []
    n_bootstrap = 1000
    
    for i in range(n_bootstrap):
        # Bootstrap sample
        X_boot, y_boot = resample(X_train, y_train, random_state=i)
        
        # Train and calculate ATE
        boot_model = clone(s_learner_model)
        fitted_boot = train_s_learner(boot_model, X_boot, y_boot)
        
        X1, X0 = prepare_data_for_s_learner_inference(X_val)
        ate = (fitted_boot.predict(X1) - fitted_boot.predict(X0)).mean()
        bootstrap_ates.append(ate)
    
    # 95% CI
    ci_lower = np.percentile(bootstrap_ates, 2.5)
    ci_upper = np.percentile(bootstrap_ates, 97.5)
    
    print(f"95% CI: [{ci_lower:.4f}, {ci_upper:.4f}]")
    
    return ci_lower, ci_upper
```

```python
s_ci = s_learner_confidence_intervals(
    s_learner_model=fitted_s_learner,
    X_train=X_train_with_T,
    y_train=y_train,
    X_val=X_val,
    y_val=y_val
)
print(s_ci)
```

## T-Learner

Select models for the t-learner. \
You will need a model for both intervention arms (T=1, T=0). \
Depending on your results, you may find that different models are better for different intervention arms. (For example, a linear model on T=0 and a tree-based model on T=1).


```python
def train_t_learner(model, X_train, y_train):
    """
    Return a fitted s-learner model on your training data
    :param model: 
    :param X_train: 
    :param y_train: 
    :return: 
    """
    return model.fit(X_train, y_train)

def t_learner_model_selection(models, metric, X_train, y_train, X_val, y_val):
    """
    Use the validation set to select the model with the best performance over the provided metric (e.g. rmse)
    Note that X_train and y_train should only include the appropriate intervention arm's data.
    
    You may select models in a more complicated way (like you would in real life). But here you can simplify.
    You must either return the fitted model or show the scores that lead to your decision.
    :param models: 
    :param metric:
    :param X_train: 
    :param y_train: 
    :param X_val: 
    :param y_val: 
    :return: 
    """
    best_score = float('inf')
    best_model = None

    for i, base_model in enumerate(models):
        model = sklearn.base.clone(base_model)                   # make a fresh copy
        model.fit(X_train, y_train)
        y_pred = model.predict(X_val)
        score   = metric(y_val, y_pred)
        print(f"Model {i} ({base_model.__class__.__name__}): "
              f"Validation score = {score:.4f}")

        if score < best_score:
            best_score = score
            best_model = model                      # this fitted copy is safe to return
    print(f"\nSelected Model: Validation score = {best_score:.4f}\n")
    return best_model
```

```python
import sklearn.ensemble
import sklearn.base
import sklearn.metrics
import numpy as np
import pandas as pd

X_train_with_t = X_train.copy()
X_train_with_t['T'] = T_train

X_val_with_t = X_val.copy()
X_val_with_t['T'] = T_val

models = [sklearn.ensemble.GradientBoostingRegressor(), sklearn.ensemble.RandomForestRegressor()]


# T = 1
X_train_1 = X_train_with_t[X_train_with_t['T'] == 1].drop(columns='T')
y_train_1 = y_train[X_train_with_t['T'] == 1]
X_val_1   = X_val_with_t[X_val_with_t['T'] == 1].drop(columns='T')
y_val_1   = y_val[X_val_with_t['T'] == 1]

fitted_t_learner_1 = t_learner_model_selection(
    models, sklearn.metrics.mean_squared_error, X_train_1, y_train_1, X_val_1, y_val_1
)

# T = 0
X_train_0 = X_train_with_t[X_train_with_t['T'] == 0].drop(columns='T')
y_train_0 = y_train[X_train_with_t['T'] == 0]
X_val_0   = X_val_with_t[X_val_with_t['T'] == 0].drop(columns='T')
y_val_0   = y_val[X_val_with_t['T'] == 0]

fitted_t_learner_0 = t_learner_model_selection(
    models, sklearn.metrics.mean_squared_error, X_train_0, y_train_0, X_val_0, y_val_0
)
```

```python
def calculate_t_learner_ate(t_learner_model1, t_learner_model0, X):
    """
    Receive t-learners and data, return the ATE inferred over X
    
    You must return or print the ate estimation on X based on the T-Learner.
    :param t_learner_model: 
    :param X1: 
    :param X0: 
    :return: 
    """

    y1_pred = t_learner_model1.predict(X)
    y0_pred = t_learner_model0.predict(X)
    ate = (y1_pred - y0_pred).mean()
    print(f"Estimated ATE (T-Learner): {ate:.4f}")
    return ate
  
```

### Q7. Using a Lasso model, check the ATE over different regularization parameters, alpha=1,2,3,4,5. \
Is this similar or different to what we saw earlier with the S-Learner as we increased regularization? If it was different, explain what causes this.

### A7:
As we increase the regularization parameter in the Lasso model, the estimated ATE from the T-learner goes up, which is the opposite of what we saw in the S-learner where it went down. This happens because these two methods handle regularization very differently.

The S-learner uses one model that includes treatment as a feature. When we increase regularization, it directly shrinks the treatment coefficient, making the estimated treatment effect smaller.

The T-learner uses two separate models - one for the treatment group and one for the control group. Since there's no treatment variable in either model, regularization doesn't directly shrink the treatment effect. Instead, it makes each model better by removing unimportant features and reducing overfitting. When both models become more accurate, the difference between them (which is our treatment effect) actually gets larger. This is why we see the ATE increase as we add more regularization in the T-learner.

```python
from sklearn.linear_model import Lasso
alphas = [0.05, 0.1, 0.2, 0.3, 0.4]
for alpha in alphas:
    t1_model = train_t_learner(Lasso(alpha=alpha), X_train.loc[T_train==1], y_train.loc[T_train==1])
    t0_model = train_t_learner(Lasso(alpha=alpha), X_train.loc[T_train==0], y_train.loc[T_train==0])
    print(calculate_t_learner_ate(t1_model, t0_model, X_val))
```

#### Add 95% confidence intervals around the point estimation. You have freedom to choose how to do this. For example, bootstraps, parametric methods (e.g. linear regression coefficient), etc.

```python
def t_learner_confidence_intervals(t_learner_model1, t_learner_model0, X_train, T_train, y_train, X_val, T_val, y_val, n_bootstrap=1000, alpha=0.05):
    """
    Receives a t-learner model for each intervention arm and returns 95% confidence intervals.
    You do not need to use every parameter passed to this method and you can add additional parameters.
    
    You must either print, visualize the confidence intervals or return them as a list.
    :param t_learner_model1: 
    :param t_learner_model0: 
    :param X_train: 
    :param y_train: 
    :param X_val: 
    :param y_val: 
    :return: 
    """
    ates = []

    for _ in range(n_bootstrap):
        X_resampled, y_resampled = sklearn.utils.resample(X_val, y_val)
        y1_pred = t_learner_model1.predict(X_resampled)
        y0_pred = t_learner_model0.predict(X_resampled)
        ate = np.mean(y1_pred - y0_pred)
        ates.append(ate)

    lower = np.percentile(ates, 100 * (alpha / 2))
    upper = np.percentile(ates, 100 * (1 - alpha / 2))
    ate_point = np.mean(t_learner_model1.predict(X_val) - t_learner_model0.predict(X_val))

    print(f"Estimated ATE (T-Learner): {ate_point:.4f}")
    print(f"95% Confidence Interval: [{lower:.4f}, {upper:.4f}]")
    return (ate_point, (lower, upper))
```

## IPW

You have already trained and selected a propensity model earlier, use that model to estimate the ATE using IPW on the validation set.

```python
def calculate_ipw_ate(model, X, T, y):
    """
    Receive propensity model and X, returns ATE inferred on X
    
    You must return or print the ate estimation on X based on IPW.
    :param model: 
    :param X: 
    :return: 
    """
    # Get propensity scores
    propensity_scores = model.predict_proba(X)[:, 1]
    
    # Calculate IPW weights
    weights = T / propensity_scores + (1 - T) / (1 - propensity_scores)
    
    # Calculate weighted outcomes for each group
    treated_outcome = np.sum(weights * T * y) / np.sum(weights * T)
    control_outcome = np.sum(weights * (1 - T) * y) / np.sum(weights * (1 - T))
    
    # Calculate ATE
    ate_ipw = treated_outcome - control_outcome
    
    print(f"IPW ATE: {ate_ipw:.4f}")
    return ate_ipw
```

```python
import numpy as np
from sklearn.base import clone
from sklearn.utils import resample

def ipw_confidence_intervals(model, X_train, T_train, y_train, X_val, T_val, y_val):
   """
   Receives a (fitted) propensity model for each intervention arm and returns 95% confidence intervals.
   You do not need to use every parameter passed to this method and you can add additional parameters.
   
   You must either print, visualize the confidence intervals or return them as a list.
   :param model: 
   :param X_train: 
   :param T_train: 
   :param y_train: 
   :param X_val: 
   :param T_val: 
   :param y_val: 
   :return: 
   """
   
   bootstrap_ates = []
   n_bootstrap = 1000
   
   for i in range(n_bootstrap):
       # Bootstrap sample from training data
       X_boot, T_boot, y_boot = resample(X_train, T_train, y_train, random_state=i)
       
       # Train propensity model on bootstrap sample
       prop_model = clone(model).fit(X_boot, T_boot)
       
       # Calculate IPW ATE on validation set
       propensity_scores = prop_model.predict_proba(X_val)[:, 1]
       weights = T_val / propensity_scores + (1 - T_val) / (1 - propensity_scores)
       
       treated_outcome = np.sum(weights * T_val * y_val) / np.sum(weights * T_val)
       control_outcome = np.sum(weights * (1 - T_val) * y_val) / np.sum(weights * (1 - T_val))
       
       ate_ipw = treated_outcome - control_outcome
       bootstrap_ates.append(ate_ipw)
   
   # Calculate confidence intervals
   ci_lower = np.percentile(bootstrap_ates, 2.5)
   ci_upper = np.percentile(bootstrap_ates, 97.5)
   original_ate = calculate_ipw_ate(model, X_val, T_val, y_val)
   
   print(f"IPW ATE: {original_ate:.4f}")
   print(f"95% CI: [{ci_lower:.4f}, {ci_upper:.4f}]")
   
   return [ci_lower, ci_upper]
```

#### Q9. Explain your matching algorithm. Are there advantages and/or disadvantages to your matching algorithm?

You will be evaluated on your explanation of the advantages and/or disadvatages of your matching algorithm. Not the algorithm itself. (Unless it is incorrect. For example, if it does not match or if it matches arbitrarily.) 


### A9:
**Advantages.1
3. **Computational speed**: Executes quickly with sklearn's optimized NearestNeighbors implementation, making it scalable for moderate datasets2
4. **Complete matching**: Guarantees every treated unit gets matched, preserving full sample size and statistical powe3.
5. **All-feature approach**: Uses all available features rather than cherry-picking, potentially capturing more confounding relationships.

**Disadvantages:**
1. **No quality thresholds**: Accepts any nearest neighbor match regardless of distance, potentially pairing very dissimilar units when good controls aren't available.
2. **Replacement allowed**: Control units can be reused multiple times, meaning one excellent control might get matched to several treated units while poor matches persist.
3. **Equal feature weighting**: Treats all standardized features as equally important for matching, which may not reflect their true relevance for confounding control.
4. **No theoretical backing**: Lacks formal statistical properties for bias reduction or variance estimation, unlike established methods with proven performance guarantees.
5. **Missing propensity integration**: Ignores treatment assignment probabilities, missing a key tool for handling selection bias effectively.

**Bottom line**: Robustechnical issues, but for serious causal inference, propensity score matching with calipers or optimal matching algorithms would perform better.


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestNeighbors
from scipy.special import logit, expit

def matching(X, T, y):
    """
    Perform 1-to-1 nearest neighbor matching using propensity scores (from logistic regression).
    Returns the matched X, T, y datasets separately for further analysis.
    
    :param X: Feature matrix (DataFrame)
    :param T: Treatment assignments (Series)
    :param y: Outcomes (Series)
    :return: matched_X, matched_T, matched_y
    """
    # 1. Estimate propensity scores using logistic regression
    logit_model = LogisticRegression(max_iter=1000)
    logit_model.fit(X, T)
    p_scores = logit_model.predict_proba(X)[:, 1]

    # 2. Convert to log-odds (logit space)
    logit_p = logit(np.clip(p_scores, 1e-6, 1 - 1e-6))

    # 3. Split by treatment
    idx_treated = np.where(T == 1)[0]
    idx_control = np.where(T == 0)[0]

    # 4. Caliper threshold
    caliper = 0.2 * logit_p.std()

    # 5. Build NN model on control group
    nn = NearestNeighbors(n_neighbors=1)
    nn.fit(logit_p[idx_control].reshape(-1, 1))

    pairs = []
    used_controls = set()
    for i_t in idx_treated:
        dist, neigh = nn.kneighbors([[logit_p[i_t]]], return_distance=True)
        j = idx_control[neigh[0][0]]
        if dist[0][0] <= caliper and j not in used_controls:
            used_controls.add(j)
            pairs.append((i_t, j))

    # 6. Extract matched indices
    matched_idx = [i for pair in pairs for i in pair]
    matched_X = X.iloc[matched_idx].reset_index(drop=True)
    matched_T = T.iloc[matched_idx].reset_index(drop=True)
    matched_y = y.iloc[matched_idx].reset_index(drop=True)

    return matched_X, matched_T, matched_y
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestNeighbors
from scipy.special import logit, expit

def matching_confidence_intervals(matching_algorithm, X_train, T_train, y_train, X_val, T_val, y_val):
   """
   Receives your matching algorithm for each intervention arm and returns 95% confidence intervals.
   You do not need to use every parameter passed to this method and you can add additional parameters.
   
   You must either print, visualize the confidence intervals or return them as a list.
   :param matching_algorithm: 
   :param X_train: 
   :param T_train: 
   :param y_train: 
   :param X_val: 
   :param T_val: 
   :param y_val: 
   :return: 
   """
   n_boot       = 100          # number of bootstrap resamples
   alpha        = 0.05          # 95 % CI
   rng          = np.random.default_rng(42)
   n_val        = len(X_val)
   boot_atts    = np.empty(n_boot)
   
   # Helper: ATT for a matched sample (simple mean difference)
   def att_from_matched(T, y):
       return y[T == 1].mean() - y[T == 0].mean()
   
   for b in range(n_boot):
       # 1) resample validation indices with replacement
       idx = rng.integers(0, n_val, n_val)
       X_b, T_b, y_b = X_val.iloc[idx], T_val.iloc[idx], y_val.iloc[idx]
       
       # 2) run matching on the bootstrap sample
       # Handle both 3 and 4 return values
       result = matching_algorithm(X_b, T_b, y_b)
       if len(result) == 4:
           mX, mT, my, _ = result  # Ignore the model
       else:
           mX, mT, my = result
       
       # 3) skip replicate if either group is missing after matching
       if (mT == 1).sum() == 0 or (mT == 0).sum() == 0:
           boot_atts[b] = np.nan
       else:
           boot_atts[b] = att_from_matched(mT, my)
   
   # Remove failed replicates
   boot_atts = boot_atts[~np.isnan(boot_atts)]
   
   # 4) percentile CI
   lower = np.percentile(boot_atts, 100 * alpha / 2)
   upper = np.percentile(boot_atts, 100 * (1 - alpha / 2))
   
   print(f"95 % CI for ATT (matching, bootstrap): [{lower:.4f}, {upper:.4f}]")
   return lower, upper
```

```python
matching_ci = matching_confidence_intervals(matching, X_train, T_train, y_train, X_val, T_val, y_val)
print(matching_ci) # if you did not show above
```
